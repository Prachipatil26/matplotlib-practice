Matplotlib is python library use for creating static , animated, and interactive plots.

Data Visualization: It is graphical representation of data to identify trend, pattern , insights.

```python
import matplotlib.pyplot as plt

#pyplot is a collection of function that make matplotlib work as MATLAB
```

```python
x=[1,2,3,4,5]
y=[10,20,25,35,45]

plt.plot(x,y) #line chart
plt.show()  
```

```python
# 2. customized chart
```

```python
plt.figure(figsize=(4,3))
plt.plot(x,y , color='blue' , marker ='o', linestyle='--' , linewidth=2, markersize=12)
plt.title("title")
plt.xlabel('name')
plt.ylabel("marks")
plt.show()
```

```python
# Advanced Line chart - multiple Lines & Legends 
```

```python
x=[1,2,3,4,5]
y1=[10,20,25,35,45]
y2=[20,30,35,45,55]
plt.plot(x,y1,label='Sales 2024') #plotting y1 data
plt.plot(x,y2,label='Sales 2025') #plotting y2 data

plt.title("YoY Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.legend()
plt.show()
```

```python
# Bar charts
```

```python
x=[1,2,3,4,5]
y=[10,20,55,45,35]
plt.bar(x,y)
plt.title('Bar Chart Example')
plt.show()
```

```python
#5 Histogram
```

```python
#used for distribution analysis
import random 
data = [random.randint(1,10) for  _ in range(500)]
plt.hist(data, bins=20)
plt.title("Histogram Example")
plt.show()
```

```python
# 6 Pie charts
#Used to show part-to-whole relationship/data
```

```python
#Used to show part-to-whole relationship/data
categories =['A','B','C','D','E']
sales = [10,20,55,35,45]

plt.pie(sales , labels = categories ,autopct= '%1.1f%%', startangle=90)
plt.title('Piechart Example ')
plt.show()
```

```python
# 7 Scatter Plot
# used to find relationship  between variables.

y1=[10,20,25,35,45]
y2=[20,30,35,45,55]

plt.scatter(y1,y2, color='red')
plt.title("Scatter Plot Example" )
plt.show()
```

```python
#8 Subplot
#used to show multiple charts in one figure

#data1 -Bar chart
categories =['Mon','Tue','Wed','Thu','Fri']
sales = [10,20,55,35,45]

#data2 - Scatter plot
y1=[10,20,25,35,45]
y2=[20,30,35,45,55]

plt.figure(figsize=(6,4))

#first plot - Bar charts
plt.subplot(1,2,1) #row,column,position
plt.bar(categories , sales)
plt.title("Daily Sales")
plt.xlabel("Week Days")
plt.ylabel("Sales")

#second plot -  scatter charts
plt.subplot(1,2,2) #row,column,position
plt.scatter(y1 ,y2)
plt.title("User Example")
plt.xlabel("User1")
plt.ylabel("User2")

plt.show()

```

```python
#9. Matplotlib with Pandas 
```

```python
import pandas as pd

data = {
     'Month' : ['Jan' , 'Feb' ,'Mar' ,'Apr'] ,
    'Sales' : [12000,11000,13000,25000]
}
df=pd.DataFrame(data)
df
```

```python
plt.bar(df['Month'],df['Sales'])
plt.title("Matplotlib with Pandas")
plt.xlabel("Month")
plt.ylabel("sales")
plt.show()
```

```python
#Save Charts
```

```python
plt.bar(df['Month'],df['Sales'])
plt.title("Matplotlib with Pandas")
plt.xlabel("Month")
plt.ylabel("sales")
plt.savefig("Monthly_Sales_from_df.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create Figure and Axes
fig, ax = plt.subplots()

# Plot using Axes object
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Customize
ax.set_title("Simple Line Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show()
```
